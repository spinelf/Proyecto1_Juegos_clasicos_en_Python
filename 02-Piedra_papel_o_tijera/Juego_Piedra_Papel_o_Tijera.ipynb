JUEGO PIEDRA, PAPEL O TIJERA

```python
import random

opciones = ("piedra", "papel", "tijera")

jugador1 = "usuario"
jugador2 = "computadora"
rondas = 1
jugador1_gana = 0
jugador2_gana = 0

print(input("Bienvenido al juego de Piedra, Papel o Tijera"))

while True:
    
    print("=" * 20)
    print("RONDA", rondas)
    print("=" * 20)
    
    print("jugador1_gana", jugador1_gana)
    print("jugador2_gana", jugador2_gana)
    
    respuesta_jugador1 = input("piedra, papel o tijera =>")
    respuesta_jugador1 = respuesta_jugador1.lower()
   
    if respuesta_jugador1 not in opciones:
        print("Esa opción no es válida")
        continue
        
    respuesta_jugador2 = random.choice(opciones)
    
    print("respuesta_jugador1 =>", respuesta_jugador1)
    print("respuesta_jugador2 =>", respuesta_jugador2)
    
    if respuesta_jugador1 == respuesta_jugador2:
        print("Empate!")
        
    elif respuesta_jugador1 == "piedra":
        if respuesta_jugador2 == "tijera":
            print("Piedra gana a tijera")
            print("Jugador1 gano!")
            jugador1_gana +=1
        
        else:
            print("Papel gana a piedra")
            print("Jugador2 gano!")
            jugador2_gana +=1
        
    elif respuesta_jugador1 == "papel":
        if respuesta_jugador2 == "piedra":
            print("Papel gana a piedra")
            print("Jugador1 gano!")
            jugador1_gana +=1
            
        else:
            print("Tijera gana a papel")
            print("Jugador2 gano!")
            jugador2_gana +=1
             
    elif respuesta_jugador1 == "tijera":
        if respuesta_jugador2 == "papel":
            print("Tijera gana papel")
            print("Jugpediraador1 gano!")
            jugador1_gana +=1
            
        else:
            print("Piedra gana a tijera")
            print("Jugador2 gano!")
            jugador2_gana +=1
            
    if jugador1_gana == 3:
        print("El ganador es el jugador1")
        break
    
    if jugador2_gana == 3:
        print("El ganador es el jugador2")
        break

    rondas +=1
       
```
